```python
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import requests
import time
import openweathermapy.core as owm
from pprint import pprint 

# Import API key
from api_keys import api_key

# Incorporated citipy to determine city based on latitude and longitude
from citipy import citipy

# Range of latitudes and longitudes
lat_range = (-90, 90)
lng_range = (-180, 180)
```

## Generate Cities List

```python
# empty cities list placeholder
cities = [] 

# Create a set of random lat and lng combinations
lats = np.random.uniform(low=-90.000, high=90.000, size=1500)
lngs = np.random.uniform(low=-180.000, high=180.000, size=1500)
lat_lngs = zip(lats, lngs)

# Identify nearest city for each lat, lng combination; city is a string -  print(city.value())
for lat_lng in lat_lngs:
    city = citipy.nearest_city(lat_lng[0], lat_lng[1]).city_name
    
    # If the city is unique, then add it to a our cities list
    if city not in cities:
        cities.append(city)

# Print the city count to confirm sufficient count
total_cities = len(cities)
print(f'Total Cities = {total_cities}')

# Uncomment the below to test the variables
# print(f"latitudes: {lats}")
# print(f"longitudes: {lngs}")
# print(f"lat_lngs is: {lat_lngs}")
```

### Perform API Calls
* Perform a weather check on each city using a series of successive API calls.
* Include a print log of each city as it'sbeing processed (with the city number and city name).


```python
# Set up the environment in order to be able to run a loop that calls each API to extract the weather data

# convert cities to a dataframe in order to manipulate the data below
cities_df = pd.DataFrame(cities, columns = ['cities']) 

# replace spaces in city names to '+' in order to work when plugged into the url
cities_no_space_df = cities_df['cities'].str.replace(' ', '+')

# convert cities_df to a dataframe
cities_no_space_df = pd.DataFrame(cities_no_space_df, columns = ['cities']) 

# counter for number of API calls
counter = 1

# error count to record how many cities the API responds back with no data.
error = 0 

# counter for the number of call sets
sets = 1 
index = 0 


url = "http://api.openweathermap.org/data/2.5/weather?"
all_cities_query_url =[]

# empty dataframe for final weather chart 
weather_df = pd.DataFrame(columns = ['City', 'Cloudiness', 'Country', 'Date', 'Humidity', 'Lat', 'Lon', 'Max Temp', 'Wind Speed'])


print(f'Beginning Data Retrieval \n ------------------------')
print(f'Go grab a cup of coffee as this will take at least 13 minutes to finish \n ------------------------')

# code breaks when variables are dataframes. They must be lists
# empty lists...
cities = [] 
city = []  
cloudiness = []
country = []
date = []
humidity = []
lat = []
lon = []
maxtemp = []
wind_speed = []



for i in range(0,total_cities):

    # query_url contains the Imperial Units (Farenheight and mph)
    query_url = url + "appid=" + api_key + "&q=" + cities_no_space_df['cities'][i] + "&units=imperial"
    all_cities_query_url.append(query_url)
    
    # get the weather data
    weather_response = requests.get(query_url)
    weather_json = weather_response.json()
    print(f'Processing Record {counter} of Set {sets} |' + cities_df['cities'][i]) 
    print(query_url)          

    try:
    # Do a print(weather_json.keys()) to see the keys 
        cloudiness.append(weather_json['clouds']['all'])
        country.append(weather_json['sys']['country']) 
        date.append(weather_json['dt'])
        humidity.append(weather_json['main']['humidity'])
        lat.append(weather_json['coord']['lat'])
        lon.append(weather_json['coord']['lon'])
        maxtemp.append(weather_json['main']['temp_max'])
        wind_speed.append(weather_json['wind']['speed'])
    except KeyError:   
        print(weather_json)   
        print('City not found. skipping...')

    error = error + 1
    counter = counter + 1
    
    # OpenWeatherMap API limit is 60 calls/min. loop to limit 50 API calls/min then pause and resume. 
    if counter == 51: 
        counter = 1 
        sets = sets + 1
        time.sleep(60)
                

print(f'Could not get data for {error} cities')
print('----------------------------------------')
print('Data Retrieval Complete')
print('----------------------------------------')

```

```python
# combine all the separate weather lists to one dataframe called weather_df
weather_df = pd.DataFrame()
weather_df['cloudiness'] = cloudiness
weather_df['country'] = country
weather_df['date'] = date
weather_df['humidity'] = humidity
weather_df['lat'] = lat
weather_df['lon'] = lon
weather_df['maxtemp'] = maxtemp
weather_df['wind_speed'] = wind_speed
```

```python
# Display the count of cities with weather data
weather_df.count()
```

#### Latitude vs. Temperature Plot

```python
# Build the scatter plots for each city types with graph properties
latitude = plt.scatter(lat, maxtemp, marker="o", facecolors="orange", edgecolors="black", alpha=0.75)

# Create a legend
plt.title("City Latitude vs. Max Temperature (08/22/18)")

plt.xlabel("Latitude")
plt.xlim(-60, 80)

plt.ylabel("Max Temperature (F)")
plt.ylim(20, 100)

plt.grid()
plt.tight_layout()

# Save Figure
plt.savefig('Lat_vs_MaxTemp', bbox_inches = 'tight', pad_inches = 2)

# Show plot
plt.show()
```

* "City Latitude vs Max Temperature" scatterplot Analysis:

In general, we can observe that the closer a city is to the equator, the higher the temperatures.  The latitudes between 0 and 40 had the higer temperatures averaging around 80 degrees Farenheight and topping at around 95 degrees. There are some outliers in the data, specifically around the 60 latitude, we can see some unusual high temperatures that occurred during that year. 

Alternatively, we can also say that there is a positive weak relationship that can be seen from -60 through 0 latitude and a negative weak relationship between latitude around 20 through 70 as there are some overlap between some of the plots.
